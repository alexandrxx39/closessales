```python
ALL = 'ALL'
def unique_sorted_values_plus_ALL(array):
    unique = array.unique().tolist()
    unique.sort()
    unique.insert(0, ALL)
    return unique
```

```python
dropdown_year = widgets.Dropdown(options =    unique_sorted_values_plus_ALL(df["Наименование ТТ"]))
output_year = widgets.Output()

def dropdown_year_eventhandler(change):
    output_year.clear_output()
    with output_year:
        if (change.new == ALL):
            display(df)
        else:
             display(df[df["Наименование ТТ"] == change.new])
dropdown_year.observe(dropdown_year_eventhandler, names='value')

display(dropdown_year)
```
